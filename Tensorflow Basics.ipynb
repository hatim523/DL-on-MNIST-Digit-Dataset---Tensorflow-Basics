```python
import tensorflow as tf
```

```python
from tensorflow.keras.datasets import mnist
```

```python
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
train_images.shape
```

```python
train_labels.shape
```

```python
test_images.shape
```

```python
from tensorflow.keras import layers
from tensorflow.keras import models
```

```python
network = models.Sequential()
```

```python
# Dense meaning one layer is connected to all other layers in next
network.add(layers.Dense(512, activation='relu', input_shape=(28*28,)))
```

```python
network.add(layers.Dense(10, activation='softmax'))
```

```python
network.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics = ['accuracy'])
```

```python
#Prepping the training data
train_images = train_images.reshape((60000, 28*28))
#scaling the data
train_images = train_images.astype(dtype='float32') / 255
```

```python
# Similiarly prepping the test data
test_images = test_images.reshape((10000, 28*28))
test_images = test_images.astype(dtype='float32') / 255
```

```python
train_images.shape
```

```python
train_images[0]
```

```python
train_labels[0]
```

```python
# ordinal -- number data (functions can be applied mean, mod, sort, conditions)
# nominal -- data on which functions cannot be applied (e.g Hatim, Hussain, Ali)
```

```python
# one hot encoding #removes biasness
from tensorflow.keras.utils import to_categorical

train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
train_labels[0]
```

```python
# batch_size --> number of data to process in one time | currently 128 images
network.fit(train_images, train_labels, epochs=30, batch_size=128)
```

```python
test_loss, test_acc = network.evaluate(test_images, test_labels)
```
